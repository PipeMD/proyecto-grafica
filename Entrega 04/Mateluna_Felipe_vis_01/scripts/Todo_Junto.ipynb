## **Primer Gráfico**

```python
import pandas as pd
import altair as alt

df = '/content/NBA ANNUAL TEAMS ATTENDANCE.csv'
nba_data = pd.read_csv(df)

teams = ["Bulls", "Mavericks", "Lakers", "Rockets", "Spurs"]
data_filtrada = nba_data[nba_data['Team'].isin(teams)][['Year', 'Team', 'Home Total Attendance']]
color_scale = alt.Scale(domain=teams, range=['#CE1141', '#00538C', '#FDB927', '#C4CED4', '#000000'])

gráfico = alt.Chart(data_filtrada).mark_line(point=True).encode(
    x='Year:O',
    y='Home Total Attendance:Q',
    color=alt.Color('Team:N', scale=color_scale),
    tooltip=['Year', 'Team', 'Home Total Attendance']
).properties(
    title='Asistencia Local Total por Año y Equipo en la NBA (Equipos Seleccionados)',
    width=1200,
    height=900
).interactive()


gráfico.save('Data Asistencia Filtrada.html')
```

## **Segundo Gráfico**


```python
import pandas as pd
import altair as alt


df = '/content/NBA ANNUAL TEAMS ATTENDANCE.csv'
nba_data = pd.read_csv(df)


team_attendance = nba_data.groupby('Team', as_index=False)['Home Total Attendance'].sum()


gráfico = alt.Chart(team_attendance).mark_bar().encode(
    x=alt.X('Home Total Attendance:Q', title='Home Total Attendance'),
    y=alt.Y('Team:N', sort='-x', title='Team'),
    tooltip=['Team', 'Home Total Attendance']

).properties(
    title='Total Home Attendance por Equipo en la NBA',
    width=800,
    height=600
)


texto = gráfico.mark_text(
    align='left',
    baseline='middle',
    dx=3
).encode(
    text='Home Total Attendance:Q'
)


gráfico_final = gráfico + texto


gráfico_final.save('Data Asistencia Completa Barras.html')
```
